Here is a view of the background information of the task:

In a Stroop task, participants are presented with a list of words, with each word displayed in a color of ink. The participantâ€™s task is to say out loud the color of the ink in which the word is printed. The task has two conditions: a congruent words condition, and an incongruent words condition. In the __congruent__ words condition, the words being displayed are color words whose names match the colors in which they are printed: for example RED, BLUE. In the __incongruent__ words condition, the words displayed are color words whose names do not match the colors in which they are printed: for example PURPLE, ORANGE. In each case, we measure the time it takes to name the ink colors in equally-sized lists. Each participant will go through and record a time from each condition.

__Question__

1) What is our independent variable? What is our dependent variable?

__Answer__

The __dependant variable__ in this case is the time taken to name the ink colors. 

The __independant variable__ is the color of the ink in which the word is printed.





__Question__

2) What is an appropriate set of hypotheses for this task? What kind of statistical test do you expect to perform? Justify your choices.

__Answer__

The null hypothesis $H_{0}$ for this task is:

$H_0 : \mu_c = \mu_I$ There is __no difference__ between the mean time in indentifying the __congrugent__ words that the mean time in identifying the __incongrugent__ words.  

The alternate hypothesis $H_A$ for this task is:

$H_A: \mu_C < \mu_I$. The mean time to identify __incongrugent__ words will take __no longer than__ the mean time taken to identify the __incrogugent__ words.

__Question__

5) Now, perform the statistical test and report your results. What is your confidence level and your critical statistic value? Do you reject the null hypothesis or fail to reject it? Come to a conclusion in terms of the experiment task. Did the results match up with your expectations?

__Answer__



There are two types of statistical test that can be conducted. Z-test and T-Test

Z-test are useful when we know the population parameters $\mu$ and $\sigma$. 

However for this data we have collected in the stroopdata.csv we just have a sample of 24 that we are looking at.

When performing the statistical test, we use the dependant T-test. The reason is that in this case the same subject/person is asked to identify Congruent and Incongruent words and in fact it makes sense when comparing that it is the same person doing so. This results in two values one for congruent words and one for incongruent words.

For performing the ttest for dependant samples on python we use scipy.stats.ttest_rel to check whether the two samples have identical average values and therefore be able to reject or not the null hypothesis.

```python
from scipy import stats
```

```python
stats.ttest_rel(df_stroop['Congruent'], df_stroop['Incongruent'])
```

Here since the p value results obtainted is much smaller than 5% (0.05) threshold, __we can reject the null hypothesis__. This 5% benchmark is the one used in most papers and examples but however for this test it is also less than 1% so in all cases can be reject the null hypothesis. This 5% or 1% threshold, is known as the __Type I error__ which is when the null hypothesis is true but it is rejected. It is like a false positive.

The result of rejecting the null hypothesis makes sense compared to what is intuitive. It makes sense it doesn't take subjects the same time in average to identify Incongruent words than congruent words.
